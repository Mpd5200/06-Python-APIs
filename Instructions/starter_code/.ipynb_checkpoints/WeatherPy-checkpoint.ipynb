```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import json
from datetime import datetime 

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy


output_data_file = "output_data/cities.csv"
```

## Generate Cities List

```python
#create a new df called cities_df to store information

cities_df = pd.DataFrame()

# Build data frame of randomly generated lat and long

cities_df['Lat'] = [np.random.uniform(-90,90) for x in range(1500)]
cities_df['Lng'] = [np.random.uniform(-180, 180) for x in range(1500)]

# add closest city and country column
cities_df['City'] = ""
cities_df['Country'] = ""

#find and add closest city and country code
for index, row in cities_df.iterrows():
    lat = row['Lat']
    lng = row['Lng']
    cities_df.set_value(index, 'City', citipy.nearest_city(lat, lng).city_name)
    cities_df.set_value(index, 'Country', citipy.nearest_city(lat, lng).country_code)

    
#remove duplictes and NA
cities_df = cities_df.drop_duplicates(['City', 'Country'])
cities_df = cities_df.dropna()

#set sample size to 500

cities_df.sample(n=500)
cities_df.head()

#check to make sure there are no duplicates
#cities_df['City'].unique
```

## Perform API Calls

```python
# Row counter to test call

row_counter = 1 

#create blank columns in the dataframe 

cities_df['Temp']=''
cities_df['Cloudiness']=''
cities_df['Humidity']=''
cities_df['Wind Speed']=''
cities_df['Date']=''


# OpenWeatherMap API Key
api_key = api_keys.api_key

#Loop through the dataframe 

for index, row in cities_df.iterrows():
    
# Starting URL for Weather Map API Call
    url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 
   
    
#create query url using api documentation 

    query_url = url + "&lat=" +str(row['Lat']) + '&lon=' + str(row["Lng"])
    print(row_counter)
    print(query_url)
    row_counter+=1

    weather = requests.get(query_url).json()
    
#create an exception handler
    try:
        temperature = weather["main"]["temp"]
        humidity = weather["main"]["humidity"]
        cloudiness = weather["clouds"]["all"]
        wind_speed = weather["wind"]["speed"]
        date = weather['dt']
        
#Add to the data frame

        cities_df.set_value(index, "Temp", temperature)
        cities_df.set_value(index,"Humidity", humidity)
        cities_df.set_value(index,"Cloudiness", cloudiness)
        cities_df.set_value(index,"Wind Speed", wind_speed)
        cities_df.set_value(index, "Date", date)
        
    except:
        print("Error with weather data, skipping.")

```

```python
#check counts
cities_df.count()
```

```python
#cities_df['City'].unique
#rename data Frame


cities_df = cities_df[[                "City",
                                       "Country",
                                       "Lat",
                                       "Lng",
                                       "Date",
                                       "Temp",
                                       "Humidity",
                                       "Cloudiness",
                                       "Wind Speed"
    
 #save as a csv
                      ]]
cities_df.to_csv("../WeatherData.csv", index=False)

cities_df
```
